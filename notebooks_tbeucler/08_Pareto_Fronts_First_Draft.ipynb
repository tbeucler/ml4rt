```python
# Scatter the loss as a function of complexity
# and label each point with the equation
fig, ax = plt.subplots(figsize=(20, 20))
ax.scatter(df['Complexity'], df['Loss'], s=siz, c='k')
for i, txt in enumerate(df['Equation']):
    ax.annotate(txt, (df['Complexity'][i], df['Loss'][i]), fontsize=20)
ax.set_xlabel('Complexity')
ax.set_ylabel('Loss')
ax.set_title('Equation Discovery')
plt.show()
```
